```python
import contextily 
ax= borough.to_crs(epsg=3857).plot(figsize=(12,6),facecolor="none",edgecolor="k")
contextily.add_basemap(ax)
```

```python
from shapely.geometry import Point
```

```python
duomo=Point(9.190680,45.464177)
```

```python
borough[borough.contains(duomo)]
```

```python
ax=borough[borough.contains(duomo)].to_crs(epsg=3857).plot(figsize=(10,8),facecolor="none",edgecolor="k")
geopandas.GeoSeries([duomo], crs='EPSG:4326').to_crs(epsg=3857).plot(ax=ax, color='red')
contextily.add_basemap(ax)
```

```python
touchesBorough=borough[borough.touches(borough[borough.contains(duomo)].geometry.squeeze())]
```

```python

ax=touchesBorough.to_crs(epsg=3857).plot(figsize=(10,8),facecolor="none",edgecolor="k")
geopandas.GeoSeries([duomo], crs='EPSG:4326').to_crs(epsg=3857).plot(ax=ax, color='red')
contextily.add_basemap(ax)
```
