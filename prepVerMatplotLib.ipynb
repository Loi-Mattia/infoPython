```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/wtitze/4E/main/prime_titles.csv")
df
```

```python
data2a = df.dropna(subset = ["Name of the show"])
data2a
```

```python
df.dropna(inplace=True) 
```

```python
df
```

```python
df1=df.groupby('Language').size().reset_index(name='counts').sort_values("counts",ascending=False)
df1
```

```python
fig,ax = plt.subplots(figsize=(30,8))
ax.bar(df1["Language"],df1["counts"])
plt.xticks(rotation=90)
plt.title("Language chart")
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(20,20))
ax.pie(df1["counts"],autopct= '%.2f%%',labels=df1["Language"])
plt.show()
```

```python
dfcopy=df.copy()
dfcopy["Language"].loc[dfcopy['Language'] !="English"] ='non English'
dfcopy
```

```python
df2=dfcopy.groupby('Language').size().reset_index(name='counts')
df2
```

```python
df3=dfcopy.groupby(['Language','Year of release']).size().reset_index(name='counts').sort_values("Year of release",ascending=False)
df3
```

```python
df4=dfcopy.groupby(['Year of release', 'Language']).size().unstack().sort_values('Year of release',ascending=False).head(10)
df4
```

```python
df4["English"]
```

```python
dfList = df4.index.tolist()
lblLoc = np.arange(len(dfList))  
width = 0.30 
fig,ax = plt.subplots()
ax.bar(lblLoc - width/2,df4["English"],width)
ax.bar(lblLoc + width/2,df4["non English"],width)
plt.xticks(lblLoc, dfList)
plt.xticks(rotation=90)
ax.legend(["English",'non English'])
plt.show()
```
