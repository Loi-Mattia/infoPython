```python
df1=df.groupby(['country']).agg({'daily_vaccinations':'sum'}).nlargest(15,"daily_vaccinations").reset_index()
df1 
```

```python
import matplotlib.pyplot as plt
df1.set_index(df1['country'],drop=True,inplace=True)
labels= df1['country']
values= df1['daily_vaccinations']
ax=df1.plot.bar(figsize=(20, 12))
plt.title('World Vaccinations')
for p in ax.patches:
    ax.annotate('{0:.0f}'.format(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005)) 
plt.bar(labels,values,color=['red', 'blue', 'purple', 'green'])
plt.show()

```
